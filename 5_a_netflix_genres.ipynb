```python
netflix = netflix.drop(columns=["ordering", "titleType", "imdb_titles"]).copy()
```

```python
netflix.dtypes
```

```python
netflix["tconst"] = np.where(netflix["tconst"].isna(), netflix["titleId"], netflix["tconst"])
```

```python
netflix
```

```python
netflix = netflix[["show", "year", "rating", "imdb", "rotten_tomatoes", "tconst"]].copy()
```

```python
netflix
```

```python
imdb = pd.read_pickle("imdb_tv_all.pkl")
```

```python
netflix = netflix.merge(imdb, how="left", on="tconst")
```

```python
netflix
```

```python
netflix["genres"].value_counts()
```

```python
netflix["genres"] = netflix["genres"].astype("str")
```

```python
genres = [genre.split(",") for genre in netflix["genres"]]
```

```python
uniques = []

for i in genres:
    for j in i:
        if j not in uniques:
            uniques.append(j)
```

```python
uniques
```

```python
uniques.remove("\\N")
```

```python
uniques.remove("nan")
```

```python
uniques
```

```python
len(uniques)
```

```python
for genre in uniques:
    netflix[genre] = 0
```

```python
netflix
```

```python
# netflix.to_pickle("netflix_genres.pkl")
```

```python
for genre in uniques:
    netflix[genre] = netflix["genres"].str.contains(genre).astype(int)
```

```python
netflix.head()
```

```python
netflix_genres = pd.DataFrame(netflix[uniques].sum().sort_values(ascending=False))
```

```python
netflix_genres
```

```python
netflix_genres["ratio"] = round((netflix_genres[0] / len(netflix)), 4)
```
