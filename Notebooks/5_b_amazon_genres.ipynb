```python
amazon["genres"] = amazon["genres"].astype("str")
```

```python
def unique_genres(df):
    
    # get list of unique genres
    genres = [genre.split(",") for genre in df["genres"]]
    uniques = []

    for i in genres:
        for j in i:
            if j not in uniques:
                uniques.append(j)
    return uniques
```

```python
uniques = unique_genres(amazon)
```

```python
uniques
```

```python
uniques.remove("\\N")
```

```python
uniques.remove("nan")
```

```python
uniques
```

```python
def genres(df, uniques):
    
    # create columns for each genre
    for genre in uniques:
        df[genre] = 0
    
    # place 1 if tv show has that genre
    for genre in uniques:
        df[genre] = df["genres"].str.contains(genre).astype(int)
        
    return df
```

```python
amazon = genres(amazon, uniques).copy()
```
